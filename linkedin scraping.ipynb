```python
Cell_no=[]
email=[]
Company=[]
comp_no=[]
Title=[]
location=[]
profile=[]
Name=[]
Reset=5#Resetting after every 5 companies. Seleium likely to throw errors otherwise

for l in range(0,len(comps),Reset):
    b=min(l+Reset,len(comps))
    comps_2=comps[l:b]
    browser= webdriver.Chrome(options=options, executable_path=DRIVER_PATH)
    browser.get("https://www.linkedin.com/")
    time.sleep(3)
    #passing Username and password
    inputelement0 = browser.find_element_by_xpath('//*[@id="session_key"]')
    inputelement1 = browser.find_element_by_xpath('//*[@id="session_password"]')
    inputelement0.send_keys(Username)
    inputelement1.send_keys(Password)
    inputelement2 = browser.find_element_by_xpath('//*[@id="main-content"]/section[1]/div/div/form/button')
    inputelement2.click()
    time.sleep(2)
    search=browser.find_element_by_css_selector('input.search-global-typeahead__input.always-show-placeholder')
    time.sleep(3)
    for i in range(len(comps_2)):
        tot=0
        time.sleep(3)
        search.clear()
        search.send_keys(String + str(comps_2[i]))
        time.sleep(2)
        search.send_keys(Keys.ENTER)
        time.sleep(7)
        # Clciking All filters
        browser.find_elements_by_css_selector("button.artdeco-pill.artdeco-pill--slate.artdeco-pill--choice.artdeco-pill--2.search-reusables__filter-pill-button.search-reusables__filter-pill-button")[-1].click()
        time.sleep(2) 
        try:
            #Clicking reset incase filter are already there 
            browser.find_elements_by_css_selector("span.artdeco-button__text")[9].click()
        except ElementClickInterceptedException or ElementNotInteractableException:
            print('No Need')
        time.sleep(2)
        #Clicking 1st Connection
        browser.find_elements_by_css_selector("span.t-14.t-black--light.t-normal")[0].click()  
        time.sleep(2)
        #Clicking Show Results 
        browser.find_elements_by_css_selector("span.artdeco-button__text")[10].click()
        time.sleep(7)

        #Filtering for current company
        browser.find_elements_by_css_selector("button[aria-label='Current company filter. Clicking this button displays all Current company filter options.']")[0].click()
        time.sleep(4)
        company=browser.find_elements_by_css_selector("input[placeholder='Add a company']")[0]
        company.click()
        #Passing the company name 
        company.send_keys(str(comps_2[i]))
        time.sleep(2)
        company.send_keys(Keys.DOWN)
        time.sleep(2)
        company.send_keys(Keys.ENTER)
        time.sleep(2)
        #Clicking Show Results button
        browser.find_elements_by_css_selector("button[aria-label='Apply current filter to show results']")[-1].click()
        time.sleep(3)
        #Getting the list of all the filtered people
        Cons=browser.find_elements_by_css_selector("span.entity-result__title-text.t-16")
        time.sleep(3)
        tot=len(Cons)
        print(str(comps_2[i]),tot)
        # Getting Contact information for each 
        if tot>0:
            for k in range(0,min(5,tot),1):
                print(k)
                browser.find_elements_by_css_selector("span.entity-result__title-text.t-16")[k].click()
                time.sleep(5)
                Title.append(browser.find_elements_by_css_selector("div.text-body-medium.break-words")[0].text)
                location.append(browser.find_elements_by_css_selector("span.text-body-small.inline.t-black--light.break-words")[0].text)
                time.sleep(2)
                Name.append(browser.find_elements_by_css_selector("h1.text-heading-xlarge.inline.t-24.v-align-middle.break-words")[0].text)
                profile.append(browser.current_url)
                browser.find_elements_by_css_selector("#top-card-text-details-contact-info")[0].click()
                time.sleep(3) 
                Company.append(str(comps_2[i]))
                if len(browser.find_elements_by_css_selector("section.pv-contact-info__contact-type.ci-phone li.pv-contact-info__ci-container.t-14"))>0:
                    Cell_no.append(browser.find_elements_by_css_selector("section.pv-contact-info__contact-type.ci-phone li.pv-contact-info__ci-container.t-14")[0].text)
                else:
                    Cell_no.append('NA')
                time.sleep(2)
                if len(browser.find_elements_by_css_selector("section.pv-contact-info__contact-type.ci-email div.pv-contact-info__ci-container.t-14"))>0:
                    email.append(browser.find_elements_by_css_selector("section.pv-contact-info__contact-type.ci-email div.pv-contact-info__ci-container.t-14")[0].text)     
                else:
                    email.append('NA')
                browser.back()
                browser.back()
                time.sleep(2)
        else:
            comp_no.append(str(comps_2[i]))
            print("No_Connections")

        print(comps_2[i])
```
